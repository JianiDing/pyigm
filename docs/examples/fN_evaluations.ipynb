```python
# imports
import imp
from scipy.signal import medfilt

from matplotlib import pyplot as plt

from astropy.table import Table

from pyigm.fN import dla as pyi_fd

pyigm_path = imp.find_module('pyigm')[1]
```

## DLAs

### Examine Sanchez-Ramirez et al. 2016  (XQ-100)

```python
tab7_fil = pyigm_path+'/data/DLA/XQ-100/sramirez16_table7.dat'
sr16_tab7 = Table.read(tab7_fil, format='ascii')
```

```python
sr16_tab7[0:3]
```

```python
plt.clf()
ax = plt.gca()
ax.plot(sr16_tab7['z'], sr16_tab7['lx'])
plt.show()
```

#### Median smooth

```python
lxmed = medfilt(sr16_tab7['lx'], 21)
plt.clf()
ax = plt.gca()
ax.plot(sr16_tab7['z'], lxmed)
plt.show()
```
