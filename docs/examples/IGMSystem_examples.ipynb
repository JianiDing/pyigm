### Generate an HI Model

```python
wave = np.arange(3800., 8000., 0.2)
xspec = XSpectrum1D.from_tuple((wave, np.ones_like(wave), np.ones_like(wave)*0.1))
```

```python
himodel, lines = pyautils.hi_model(hisys, xspec, add_lls=True)
```
